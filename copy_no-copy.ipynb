# Subarrays as no-copy views

```python
import numpy as np
```

```python
x2 = np.array([[12,5,2,4],[7,6,8,8],[1,6,7,7]])
x2
```

```python
x2_sub = x2[:2,:2]
x2_sub
```


### Here x2 and x2_sub have the same location for the same array because x2_sub holds sub array from x2 

```python
x2_sub[0,0] = 99 # we change the first element in x2_sub will affect the original x2
x2_sub
```

```python
x2
```

# Creating copies of arrays

```python
x2_sub_copy = x2[:2,:2].copy()
x2_sub_copy
```

```python
x2_sub_copy[0,0] = 42
x2_sub_copy
```

```python
x2 # the original one doesn't change
```
